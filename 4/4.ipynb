３０．形態素解析結果の読み込み

```python

result = []
sentence = []
 
with open("neko.txt.mecab") as f:
  for line in f:
    l1 = line.split("\t")
    if len(l1) == 2:
      l2 = l1[1].split(",")
      sentence.append({"surface": l1[0], "base": l2[6], "pos": l2[0], "pos1": l2[1]})
      if l2[1] == "句点":
        result.append(sentence)
        sentence = []
 
result[:2]
```

３１．動詞

```python
se = set()
 
for lis in result:
  for dic in lis:
    if dic["pos"] == "動詞":
      se.add(dic["surface"])
 
print(se)
```

３２．動詞の基本形

```python
se = set()
 
for lis in result:
  for dic in lis:
    if dic["pos"] == "動詞":
      se.add(dic["base"])
 
print(se)
```

３３．AのB

```python
se = set()
 
for line in result:
  for i in range(len(line)):
   if line[i]["pos"] == "名詞" and line[i + 1]["surface"] == "の" and line[i + 2]["pos"] == "名詞":
     se.add(line[i]["surface"] + line[i + 1]["surface"] + line[i + 2]["surface"])
 
print(se)
```

３４．名詞の連接

```python
text_d = {}
text = ""
 
for line in result:
  for dic in line:
    if dic["pos"] == "名詞":
      text += dic["surface"]
    else:
      text_d[text] = len(text)
      text = ""
 
print([k for k, v in text_d.items() if v == max(text_d.values())])
```

３５．単語の出現頻度

```python
from collections import Counter
 
result = []
sentence = []
 
with open("neko.txt.mecab") as f:
  for line in f:
    l1 = line.split("\t")
    if len(l1) == 2:
      l2 = l1[1].split(",")
      sentence.append({"surface": l1[0], "base": l2[6], "pos": l2[0], "pos1": l2[1]})
      if l2[1] == "句点":
        result.append(sentence)
        sentence = []
 
text_l = []
 
for lis in result:
  for dic in lis:
    if dic["surface"] != "":
      text_l.append(dic["surface"])
 
c35 = Counter(text_l)
print(c35.most_common())
```

３６．頻度上位１０語

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
 
#count
c36 = Counter(text_l)
word, count = zip(*c36.most_common(10))
 
#plot
fig, ax = plt.subplots()
ax.bar(word, count)
plt.show()
```

３７．「猫」と共起頻度の高い上位１０語

```python
text_l_37 = []

for lis in result:
    if any(d["base"] == "猫" for d in lis):
        for dic in lis:
            text_l_37.append(dic["surface"])

c_37 = Counter(text_l_37)
word, count = zip(*c_37.most_common(10))

fig, ax = plt.subplots()
ax.bar(word, count)
plt.show()
```

３８．ヒストグラム

```python
c38 = Counter(text_l)
word, count = zip(*c38.most_common())

fig, ax = plt.subplots()
ax.hist(count, range(1, 10))
plt.xlabel("出現頻度")
plt.ylabel("種類数")
plt.show()
```

３９．Zipfの法則

```python
c39 = Counter(text_l)
word, count = zip(*c39.most_common())

x = count
y = ([i + 1 for i, v in enumerate(count)])
fig, ax = plt.subplots()
ax.plot(x,y)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
