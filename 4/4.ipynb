３０．形態素解析結果の読み込み

```python

result = []
sentence = []
 
with open("neko.txt.mecab") as f:
  for line in f:
    l1 = line.split("\t")
    if len(l1) == 2:
      l2 = l1[1].split(",")
      sentence.append({"surface": l1[0], "base": l2[6], "pos": l2[0], "pos1": l2[1]})
      if l2[1] == "句点":
        result.append(sentence)
        sentence = []
 
result
```

３１．動詞

```python
se = set()
 
for lis in result:
  for dic in lis:
    if dic["pos"] == "動詞":
      se.add(dic["surface"])
 
print(se)
```

３２．動詞の基本形

```python
se = set()
 
for lis in result:
  for dic in lis:
    if dic["pos"] == "動詞":
      se.add(dic["base"])
 
print(se)
```

３３．AのB

```python
se = set()
 
for line in result:
  for i in range(len(line)):
   if line[i]["pos"] == "名詞" and line[i + 1]["surface"] == "の" and line[i + 2]["pos"] == "名詞":
     se.add(line[i]["surface"] + line[i + 1]["surface"] + line[i + 2]["surface"])
 
print(se)
```

３４．名刺の連接

```python
text_d = {}
text = ""
 
for line in result:
  for dic in line:
    if dic["pos"] == "名詞":
      text += dic["surface"]
    else:
      text_d[text] = len(text)
      text = ""
 
print([k for k, v in text_d.items() if v == max(text_d.values())])
```
